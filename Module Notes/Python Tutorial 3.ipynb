### Error checking

```python
# We want to pass two lists, and return a list that contains the element-wise summation of the two lists

def elementSum(L1, L2):
    if len(L1) != len(L2):
        raise ValueError("The lists must be the same length")
    
    ret = []      # start with empty list
    
    for i in range(0,len(L1)):
        ret.append(L1[i] + L2[i])
    
    return ret
```

# Plotting - matplotlib

```python
import matplotlib.pyplot as plt
```

```python
L1 = [1, 1.5, 2.5, 4]

plt.plot(L1)
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])   # Similar to "scatter plot" in excel
```

```python
L1 = [1,3,6,9,15]           # Multiple plots in one graph
L2 = [2,4,6,8,10]
L3 = elementSum(L1,L2)

plt.plot(L1)
plt.plot(L2)
plt.plot(L3)
plt.show()           # Note the auto colouring
```

```python
L1 = [1,3,6,9,15]           # Multiple plots in one graph
L2 = [2,4,6,8,10]
L3 = elementSum(L1,L2)

plt.plot(L1,'g' )
plt.plot(L2,'r--')    # Change the colour and style of my plots
plt.plot(L3,'bo-')

plt.ylabel('some numbers')       # Add some labels
plt.xlabel('some other numbers')

plt.show()
```

```python
L1 = [1,3,6,9,15]           # Multiple plots in one graph
L2 = [2,4,6,8,10]
L3 = elementSum(L1,L2)

plt.plot(L1,'g', label="list 1")       # Add labels to each set of data added
plt.plot(L2,'r--', label="list 2")
plt.plot(L3,'bo-', label="sum of lists")

plt.ylabel('some numbers')
plt.xlabel('some other numbers')

plt.legend()   # call the legend function

plt.show()
```

## Numpy

You have coded by hand some calculus and linear algebra, but you will never do that in reality in Python. You will use the powerful numpy library.

```python
import numpy as np                # are you comfortable with this?

# Let's make a vector
v = np.array([1,2,3])             # Need to explain this notation - to access stuff in np, you need to use the .

print(v)
```

```python
# Let's make a matrix
A = np.array([[1,2,3],
              [2,3,4],
              [4,5,6]])

print(A)
```

```python
# Let's multiply a matrix by a vector

a = A.dot(v)

print(a)
```

```python
# Let's get another matrix
B = np.array([[2,3,4],
              [3,4,5],
              [4,5,6]])

R1 = A.dot(B)

print(R1)

R2 = B.dot(A)

print(R2)
```
